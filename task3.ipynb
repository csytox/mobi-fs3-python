## Linear filtering by convolution (mean filter)

**<a id="#">Task 3.1.1.</a>** Load the image `data/lena.png` and show it:

**<a id="#">Task 3.1.2.</a>** Implement a mean filter:

```python
img_1 = img
def meanfilter(img, size):
    result = numpy.zeros(img.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img.shape):
        neighborhood_sum = 0
        for q in numpy.ndindex(size, size):
            if p[0]<n or p[1]<n or p[0]>img.shape[0]-n-1 or p[1]>img.shape[1]-n-1:
                result[p] = 0
            else:
                intensity = img[p[0] + q[0] - n, p[1] + q[1] - n]
                neighborhood_sum += intensity           
            
        result[p] = neighborhood_sum / 9

    return result


img_2 = meanfilter(img, 3)


```

## Non-linear filtering

**<a id="#">Task 3.2.1.</a>** Implement a median filter:

```python
def medianfilter(img, size):
    result = numpy.zeros(img.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img.shape):
        neighborhood_med = 0
        l=list()
        for q in numpy.ndindex(size, size):
            if p[0]<n or p[1]<n or p[0]>img.shape[0]-n-1 or p[1]>img.shape[1]-n-1:
                result[p] = 0
            else:
                intensity = img[p[0] + q[0] - n, p[1] + q[1] - n]
                l.append(intensity)
                l.sort()
                median = l[len(l) //2]
                neighborhood_med =  median     
        result[p] = neighborhood_med
        l.clear()
    return result
```
